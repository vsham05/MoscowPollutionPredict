```python
 import pandas as pd
 import numpy as np
```

# Объединим датасеты


```python
data = []

for i in range(15, 21):
    df = pd.read_excel(f'Останкино 0 метео 20{i}.xlsx')
    data.append(df)

res_df = pd.concat(data)
```

```python
res_df
```

```python
res_df.to_csv('all_data.csv', index=False)
```

```python
pollution_data = pd.concat([pd.read_excel('Останкино 2015-2020гг.xlsx', str(i)) for i in range(2015, 2021)])
```

```python
pollution_data
```

```python
main_df = pd.read_csv('all_data.csv')
```

```python
main_df['CO'] = pollution_data['CO'].values
main_df['NO2'] = pollution_data['NO2'].values
main_df['NO'] = pollution_data['NO'].values
main_df['PM10'] = pollution_data['PM10'].values
main_df['PM2.5'] = pollution_data['PM2.5'].values
```

```python
main_df
```

Сохраним итоговый датасет:

```python
main_df.to_csv('mainData.csv')
```
